```
# 라이브러리 임포트
import os 
import re
import pandas as pd
import numpy as np

root_dir = '/content/gdrive/My Drive/ITWILL/Untitle/'
csv_list = os.listdir(root_dir + 'melon/')
print(csv_list)
```

```
number = sorted([int(i[8:-4]) for i in csv_list])
csv_list = ['singdata' + str(i) + '.csv' for i in number]
print(csv_list)
```

```
df = pd.read_csv(root_dir + 'melon/' + csv_list[0], header = None)

df.head()
```

```
# 데이터 전처리 및 저장
cnt = 0
for csv_name in csv_list :
  cnt += 1
  print(str(cnt) + '.' + csv_name + 'is porcessed')

  df = pd.read_csv(root_dir + 'melon/' + csv_name, names = ['ranking', 'title', 'artist', 'album', 'likes', 'date', 'lyrics'])

  for i in range(df.shape[0]) :
    df.loc[i, 'artist'] = df.loc[i, 'artist'][:int(len(df.loc[i, 'artist'])/2)]
    df.loc[i, 'date'] = df.loc[i, 'date'][4:-5]

  if csv_name == csv_list[0] :
    df.to_csv(root_dir + "data.csv", index = False)
  else : 
    df.to_csv(root_dir + "data.csv", mode = 'a', header = False, index = False)
```
