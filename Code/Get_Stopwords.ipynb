```python
import os
import re
import sys
import numpy as np
import pandas as pd
from konlpy.tag import Okt
```

```python
df = pd.read_csv('data_10000.csv')

df.head()
```

```python
df_ = df['lyrics']
print('Before filtering : ', len(df_))

index = []

for i in range(len(df_)) :
    if df_.iloc[i] == '가사가 없습니다' : index.append(i)

df_.drop(index, inplace = True)
print('After filtering : ', len(df_))
```

```python
df_.head()
```

```python
text = ''

for i in range(int(len(df_))) :
    text += df_.iloc[i]
    
text[:100]
```

```python
text = re.sub('<br/>', '##', text)
text = re.sub('[A-Za-z]', '', text)
text = re.sub('\n', '', text)
text = re.sub('#+', '##', text)
text = re.sub('’', '', text)
text = re.sub('##', ' br ', text)
text = re.sub(' +', ' ', text)
text = re.sub('[br ]{2,}', ' br ', text)
text = re.sub('br', '\n', text)

text[:100]
```

```python
okt = Okt()

pos = okt.pos(text)
```

```python
okt.tagset
```

```python
Foreign = []
KoreanParticle = []
Punctuation = []
Unknown = []

for i in pos :
    if i[1] == 'KoreanParticle' : KoreanParticle.append(i[0])
    elif i[1] == 'Foreign' : Foreign.append(i[0])
    elif i[1] == 'Punctuation' : Punctuation.append(i[0])
    elif i[1] == 'Unknown' : Unknown.append(i[0])
```

```python
print("외국어 및 한자 : {0}\n".format(len(set(Foreign))), set(Foreign))
print("-"*150)
print("한글조각 : {0}\n".format(len(set(KoreanParticle))), set(KoreanParticle))
print("-"*150)
print("구두점 : {0}\n".format(len(set(Punctuation))), set(Punctuation))
print("-"*150)
print("미등록어 : {0}\n".format(len(set(Unknown))), set(Unknown))
```

```python
L = []

L.extend(list(set(Foreign)))
L.extend(list(set(KoreanParticle)))
L.extend(list(set(Punctuation)))
L.extend(list(set(Unknown)))

L[:5]
```

```python
df = pd.DataFrame(L)

df.head()
```

```python
df.to_csv('stopword.csv', index = False, header = None)
```
